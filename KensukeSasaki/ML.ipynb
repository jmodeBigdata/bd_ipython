#  目的は、与えられたデータからxからyへの写像を見つけること

#教師なし学習
#  ラベルが与えられない
#  目的は、データの背後に隠れている構造を発見することである

```python
import numpy as np
import scipy.stats as st
import sklearn.linear_model as lm
import matplotlib.pyplot as plt
%matplotlib inline
f = lambda x : np.exp(3*x)
#モデルの背後にある決定論的非線形関数を定義する
```

```python
x_tr = np.linspace(0., 2, 200) #訓練データのx軸
y_tr = f(x_tr)#訓練データのy軸
#区間[0,2]の間の値を生成する
```

```python
x = np.array([0,.1,.2,.5,.8,.9,.1])
y = f(x) + np.random.randn(len(x))
#区間[0,1]の間で、データポイントを生成する。関数fを用いた上で、ガウスノイズを載せる
```

```python
plt.plot(x_tr[:100],y_tr[:100],'--k')
plt.plot(x,y,'ok',ms=10)
#[0,1]で生成したデータポイントを描画する
```

```python
# We create the model.
lr = lm.LinearRegression() #モデル作成
# We train the model on our training dataset.
lr.fit(x[:, np.newaxis],y) #訓練データを使って学習を行う
#Now, we predict points with our trained model.
y_lr = lr.predict(x_tr[:,np.newaxis])#学習済みモデルで予測を行う 

#観測を行、特徴量を列とするskikit-learnの慣習に基づき、xとx_trの列ベクトルの変換*
#(x[:,np.newaxis])が必要
#ここでは一つの特徴量に対して観測値が7つ存在
```

```python
#学習済み線形モデルを表示(回帰モデルを緑で)
plt.plot(x_tr,y_tr,'--k')
plt.plot(x_tr,y_lr,'g')
plt.plot(x,y,'ok',ms=10) #実測値、msは●の大きさ
plt.xlim(0,1)
plt.ylim(y.min()-1,y.max()+1)
plt.title("Linear regression don't mind")
```

```python
Irp = lm.LinearRegression()
plt.plot(x_tr,y_tr,'--k')
for deg in [2,5]:
    lrp.fit(np.vander(x,deg + 1),y)
    y_lrp = lrp.predict(np.vender(x_tr,deg + 1))
    plt.plot(x_tr,y_lrp,
             label = 'degree' + str(deg))
    plt.legend(loc=2)
    plt.xlim(0,1.4)
    plt.ylim(-10,40)
    # print the model's coefficients. #モデルの係数を表示
    print(' '.join(['%.2f' % c for c in lrp.coef_]))
plt.plot(x,y,'ok',ms=10)
plt.title("Linear regression")
    
```
